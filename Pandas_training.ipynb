## Making changes to data

```python
#making new column
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(5)
```

```python
#droping column
df=df.drop(columns=['Total'])
print(df)
```

```python
df['Total'] = df.iloc[:,4:10].sum(axis=1) #also adding column
df.head(5)
```

```python
#changin order of column
cols = list(df.columns.values)#takomg mames of columns
df=df[cols[0:4] + [cols[-1]] + cols[4:-2]]
df.head(5)
```

# Filtering DATA

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]

new_df.reset_index(drop = True, inplace = True) #drop means it drops old index
new_df
```

```python
new_df = df.loc[~df['Name'].str.contains('Mega')] #(~nie)zawieta slowo mega
new_df
```

```python
import re #lib to even more complex filtering
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex = True)] #flags re.I means capital letters dsnt matter

```

```python
df.loc[df['Name'].str.contains('^pi[a-z]', flags=re.I, regex = True)] #^-first letter
```

# Aggregate Statisctics (Groupby)

```python
df=pd.read_csv('modified.csv')
df.groupby(['Type 1']).mean().sort_values('HP', ascending = False)#policz srednia po dla Type1 malejaco defensywa
```

```python
df['count']=1 #dodanie pomocniczej kolumny
df
df.groupby(['Type 1', 'Type 2']).count()['count'] #pogrupowanie po typie 1 i 2 i zliczenie
```
