```python
import pandas as pd

df = pd.read_csv(
    "https://data.heatonresearch.com/data/t81-558/jh-simple-dataset.csv",
    na_values=['NA','?'])

pd.set_option('display.max_columns', 20)
pd.set_option('display.max_rows',20)


display(df)
```

## first thing is to change non numeric values into numercic

### non numeric is job and area and we can change them into dummies

### also need to get rid off of  id becouse it is not column that has influence on result


```python
df = pd.concat([df,pd.get_dummies(df['job'],prefix='job')],axis=1)
df.drop('job',axis=1,inplace=True)
display(df[0:10])
```

```python
df = pd.concat([df,pd.get_dummies(df['area'],prefix='area')],axis=1)
df.drop('area',axis=1,inplace=True)
df.drop('id',axis=1,inplace=True)
display(df[0:10])
```

```python
#becouse there is void values in income
med = df['income'].median()
df['income']=df['income'].fillna(med)
```
